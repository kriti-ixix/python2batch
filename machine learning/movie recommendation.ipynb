```python
features = ['keywords', 'cast', 'genres', 'director']
```

```python
def combineFeatures(row):
    return row['keywords'] + " " + row['cast'] + " " + row['genres'] + " " + row['director']
```

```python
for i in features:
    df[i].fillna(' ', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['combined_features'] = df.apply(combineFeatures, axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer()
```

```python
countMatrix = cv.fit_transform(df['combined_features'])
```

```python
countMatrix
```

```python
cosineSim = cosine_similarity(countMatrix)
```

```python
cosineSim
```

```python
list(cosineSim[10])
```

```python
def getIndex(title):
    return df[df['title'] == title]['index'].values[0]
```

```python
def getTitle(index):
    return df[df['index'] == index]['title'].values[0]
```

```python
getIndex('Batman')
```

```python
similarMovies = list(enumerate(cosineSim[1359]))
similarMovies
```

```python
sortedMovies = sorted(similarMovies, key = lambda x : x[1],reverse = True)[1:]
sortedMovies
```

```python
df.iloc[2793]
```

```python
getTitle(2793)
```

```python
count = 1

for movie in sortedMovies:
    index = movie[0]
    print(getTitle(index))
    
    if count == 10:
        break
        
    count += 1
```

```python
userMovieLikes = input("Enter a movie you like: ")
userIndex = getIndex(userMovieLikes)

similarMovies = list(enumerate(cosineSim[userIndex]))
sortedMovies = sorted(similarMovies, key = lambda x : x[1],reverse = True)[1:]

count = 1

for movie in sortedMovies:
    index = movie[0]
    print(getTitle(index))
    
    if count == 10:
        break
        
    count += 1
```
