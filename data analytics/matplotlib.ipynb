```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x**2
y
```

## Functional Method

```python
plt.plot(x, y)
#plt.show()
```

```python
plt.plot(x, y, 'y')
plt.title('This is my graph')
plt.xlabel('X Array')
plt.ylabel('Y Array')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'g--')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'r')
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'y')

plt.subplot(2, 2, 2)

plt.subplot(2, 2, 3)
plt.plot(y, x, 'g')

plt.subplot(2, 2, 4)

plt.tight_layout()
```

## Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'b')
axes.set_title("Title")
#add_axes(left, bottom, height, width)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes2 = fig.add_axes([0.15, 0.55, 0.3, 0.3])

axes1.plot(x, y, 'r', label = "Red")
axes2.plot(x, y, 'g', label = "Green")

axes1.legend(loc=1)
axes2.legend()
```
